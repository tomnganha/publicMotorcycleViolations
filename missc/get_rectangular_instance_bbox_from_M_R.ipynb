```python
def riders_in_motorcycle(rider, motorcycle):
    x1, x2, y1, y2 = motorcycle['x'] - motorcycle['w']/2, motorcycle['x'] + motorcycle['w']/2, motorcycle['y'] - motorcycle['h']/2, motorcycle['y'] + motorcycle['h']/2
    rider['x1'], rider['x2'], rider['y1'], rider['y2'] = rider['x'] - rider['w']/2, rider['x'] + rider['w']/2, rider['y'] - rider['h']/2, rider['y'] + rider['h']/2
    if(x1==0):
        x1 = 0.001
    rider[['fx1', 'fx2']] = (rider[['x1', 'x2']] / [x1, x2]).astype('int32')
    
    subset = rider.loc[(rider['fx1']!=0 ) & (rider['fx2']==0 )]
    if(subset.empty):
        print("No rider found")
        x=[round((x1+x2)/2,6), round((y1+y2)/2,6), round((x2-x1),6), round((y2-y1),6)]
        return x
    
    print(max(subset[['y2']].values))
    xmin = min(np.array([x1]), min(subset[['x1']].values))[0]
    xmax = max(np.array([x2]), max(subset[['x2']].values))[0]
    ymin = min(np.array([y1]), min(subset[['y1']].values))[0]
    ymax = max(np.array([y2]), max(subset[['y2']].values))[0]
    print(ymax)
    x=[(xmin+xmax)/2, (ymin+ymax)/2, (xmax-xmin), (ymax-ymin)]
    return x
```

```python
count = 0
```

```python
for files in glob.glob(folder_path + "/*.txt"):
    file = os.path.join(result_folder_path,basename(files))
    data = pd.read_csv(files, header=None, sep=" ", names=['class_id', 'x', 'y', 'w', 'h'])
    motorcycle = data.loc[(data['class_id']==3)]
    rider = data.loc[(data['class_id']==0)]
    if(len(motorcycle)==0):
        open(file,'a')
    for j in range(len(motorcycle)):
        subset = riders_in_motorcycle(rider, motorcycle.iloc[j])
        #print(j)
        #print(subset)
        with open(file,'a') as f:
            f.write('0' + ' ' + str(subset[0]) + ' ' + str(subset[1]) + ' ' + str(subset[2]) + ' ' + str(subset[3]))
            f.write('\n')
    
    df = pd.read_csv(file, header=None, sep=" ", names=['class_id', 'x', 'y', 'w', 'h'])
    if(len(df) == len(motorcycle)):
        count = count + 1
        #print("True")
```
